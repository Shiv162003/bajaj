```python
import pandas as pd 
import json
from pandas import json_normalize
```

```python
# i am trying to convert the folded jason file into a flat csv file
with open('DataEngineeringQ2.json', 'r') as json_file:
    data = json.load(json_file)

df = json_normalize(data)

df.to_csv('DataEngineeringQ2.csv', index=False)

print("Nested JSON data converted")
```

```python

def is_valid_phone_number(phone):
    if isinstance(phone, str):
        if phone.startswith('+91') and len(phone) == 13 and phone[1:].isdigit():
            return True
        elif phone.startswith('91') and len(phone) == 12 and phone.isdigit():
            return True
        elif len(phone) == 10 and phone.isdigit():
            return True
    return False

# Count invalid phone numbers
invalid_phone_count = 0
for phone in df['phoneNumber']:
    if not is_valid_phone_number(phone):
        invalid_phone_count += 1

print(f'Total number of invalid phone numbers: {invalid_phone_count}')
```
